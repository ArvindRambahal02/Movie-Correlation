```python
# import libaries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

# matplotlib configurations
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12, 8) # Adjusts the configuration of the plots we will create


# Read in the data
df = pd.read_csv(r"C:\portfolioProject\movies.csv")
```

```python
# Let's look at the data
df.head()
```

```python
# remove any missing or NAN values in dataset
df = df.dropna()
```

```python
# Determine of there is any missing data
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print("{} - {}%".format(col, pct_missing))
```

```python
# Data types for our columns
df.dtypes
```

```python
# Changing data types of columns from float to integer
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
df['votes'] = df['votes'].astype('int64')
```

```python
df.head()
```

```python
df['released'].astype(str)
```

```python
# Create correct year column

import re

def extract_year(date_string):
    match = re.search(r'\b\d{4}\b', date_string)
    if match:
        return match.group()
    else:
        return None

df['year_corrected'] = df['released'].apply(extract_year)
```

```python
df
```

```python
# Look at all of the data
#pd.set_option('display.max_rows', None) # all rows in the dataset will be displayed

# Order data by gross revenue
df = df.sort_values(by=['gross'], inplace = False, ascending = False)
```

```python
# Drop duplicate data
# use .drop_duplicates() on data frame or specific column in data set

# df.drop_duplicates() or df['column_name'].drop_duplicates() for example
df['company'].drop_duplicates()
```

```python
df.head()
```

```python
# Finding correlation in the Data

# Budget high correlation - the more money they spend the more money they would bring in
# Company high correlation
```

```python
# Scatterplot with budget vs. gross revenue

plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs Gross Earnings')

plt.xlabel('Gross Earnings')
plt.ylabel('Budget for Film')
plt.show()
```

```python
# Regresion plot
# plot budget vs gross using seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={'color': 'red'}, line_kws={'color':'blue'})
```

```python
# Correlation
#df.corr(method='pearson') # Pearson, kendall, spearman
columns_of_interest = ['budget', 'gross', 'runtime', 'score', 'votes', 'year']
df[columns_of_interest].corr(method='pearson')
```

```python
# High correlation between budget and gross
```

```python
# Visualize correlation matrix
correlation_matrix = df[columns_of_interest].corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movies Features')

plt.show()
```
